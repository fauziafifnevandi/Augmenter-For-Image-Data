```python
import Augmentor
import os

s = "C:/Users/Nyx/skripsi/cnn_fauzi/default_model/slice_wajah/wajah/output/all_bukanwajah"
a = os.listdir(s)

for i in a:
    source_dir = s 
    output_dir = "C:/Users/Nyx/skripsi/cnn_fauzi/default_model/slice_wajah/wajah/output/all_bukanwajah/output/"
    p = Augmentor.Pipeline(source_directory=source_dir, output_directory=output_dir)
#   p.random_distortion(probability=1, grid_width=4, grid_height=4, magnitude=6)
    p.rotate(probability=0.7, max_left_rotation=13, max_right_rotation=13)
 #   p.zoom_random(probability=0.5, percentage_area=0.9)
    p.crop_random(probability=0.6, percentage_area=0.9)
    p.resize(probability=1.0, width=128, height=128)

    p.sample(10)
```
